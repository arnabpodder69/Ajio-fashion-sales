```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    "Product_URL": ["url1", "url2", "url3", "url4", "url5"],
    "Brand": ["Brand1", "Brand2", "Brand1", "Brand3", "Brand2"],
    "Description": ["Description1", "Description2", "Description3", "Description4", "Description5"],
    "Id_Product": [1, 2, 3, 4, 5],
    "Category_by_gender": ["Men", "Women", "Men", "Unisex", "Women"],
    "Discount_Price (in Rs.)": [100, 150, 200, 120, 180],
    "Original_Price (in Rs.)": [120, 180, 250, 150, 220],
    "Color": ["Red", "Blue", "Green", "Black", "White"]
}

# Create DataFrame
df = pd.DataFrame(data)

# Visualization Graph (Bar Chart)
plt.figure(figsize=(10, 6))
df['Brand'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Product Count by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization Pie Chart
plt.figure(figsize=(8, 8))
df['Category_by_gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Distribution of Products by Category')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python


# Visualization 1: Bar graph - Category by gender
plt.figure(figsize=(10, 6))
df['Category_by_gender'].value_counts().plot(kind='bar')
plt.title('Number of Products by Gender Category')
plt.xlabel('Gender Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Visualization 2: Pie chart - Category by gender
plt.figure(figsize=(8, 8))
df['Category_by_gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightblue', 'lightcoral'])
plt.title('Distribution of Products by Gender Category')
plt.ylabel('')
plt.show()

# Visualization 3: Bar graph - Top 10 Brands
plt.figure(figsize=(10, 6))
top_brands = df['Brand'].value_counts().head(10)
top_brands.plot(kind='bar')
plt.title('Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Visualization 4: Pie chart - Top 10 Brands
plt.figure(figsize=(8, 8))
top_brands.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Top 10 Brands')
plt.ylabel('')
plt.show()

# Visualization 5: Bar graph - Top 10 Colors
plt.figure(figsize=(10, 6))
top_colors = df['Color'].value_counts().head(10)
top_colors.plot(kind='bar')
plt.title('Top 10 Colors')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Visualization 6: Pie chart - Top 10 Colors
plt.figure(figsize=(8, 8))
top_colors.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Top 10 Colors')
plt.ylabel('')
plt.show()

# You can continue creating more visualizations following a similar pattern for other features in your dataset.
```
